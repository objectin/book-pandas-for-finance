```python
import matplotlib.pyplot as plt

fir, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [2, 1, 1, 2])

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('title')

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], label='ascending')
plt.plot([1, 2, 3, 4], [4, 3, 2, 1], label='descending')

plt.legend(loc='best')
plt.show()
```
