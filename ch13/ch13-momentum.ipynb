## 13.2 유가증권시장 12개월 모멘텀

최근 투자 기간 기준으로 12개월 모멘텀 계산 날짜 구하기

```python
from pykrx import stock
import FinanceDataReader as fdr
```

```python
df = fdr.DataReader(symbol='KS11', start="2019-10")
start = df.loc["2019-10"]
end = df.loc["2020-09"]
```

```python
start_date = start.index[0]
end_date = end.index[-1]
print(start_date, end_date)
```

가격 모멘텀 계산 시작일 기준으로 등락률을 계산합니다. 

```python
kospi = stock.get_market_price_change_by_ticker(
    fromdate="20191002",   # 2019년 10월 01일 종가 
    todate="20200929",
    market="KOSPI"
)
kospi
```

12개월 등락률(가격 모멘텀, 최근 1개월 제외)을 기준으로 상위 20종목의 종목 코드를 가져와봅시다. 

```python
kospi.sort_values(by='등락률', ascending=False)[:20]
```

```python
kospi_momentum20 = kospi.sort_values(by='등락률', ascending=False)[:20]
kospi_momentum20 = kospi_momentum20[['종목명', '등락률']]
kospi_momentum20.rename(columns={"등락률": "Momentum"}, inplace=True)
kospi_momentum20
```

```python
pct_df = stock.get_market_price_change_by_ticker(
    fromdate="20201006",  # 2020-10-05일 종가가 시가 컬럼임
    todate="20210430",
    market="KOSPI"
)
pct_df
```

```python
pct_df = pct_df[['시가', '종가', '등락률']]
```

```python
kospi_momentum20_result = kospi_momentum20.join(pct_df)
kospi_momentum20_result
```

```python
kospi_momentum20_result['원금'] = 10000000
kospi_momentum20_result['매도금'] = (kospi_momentum20_result['원금'] * (100 + kospi_momentum20_result['등락률'])) / 100
```

```python
kospi_momentum20_result.head()
```

```python
최초자산 = kospi_momentum20_result['원금'].sum()
최종자산 = kospi_momentum20_result['매도금'].sum()
mom20_cagr = (최종자산/최초자산) ** (1/0.5) - 1    # 6개월
mom20_cagr * 100
```

```python
df_ref = fdr.DataReader(
    symbol='KS11', 
    start="2020-10-05",
    end="2021-04-30"
)
df_ref
```

```python
CAGR = ((df_ref['Close'].iloc[-1] / df_ref['Close'].iloc[0]) ** (1/0.5)) -1
CAGR * 100
```

## 13.3 대형주 12개월 모멘텀

대형주(시가총액 200위) 기준으로 상대 모멘텀이 큰 종목을 20개 선정하는 전략

```python
all = stock.get_market_price_change_by_ticker(
    fromdate="20191002",   # 2019년 10월 01일 종가 
    todate="20200929",
    market="all"
)
all
```

```python
# 기본 필터링
# 우선주 제외 
all2 = all.filter(regex="0$", axis=0)
```

```python
all2 = all2[['종목명', '등락률']]
all2
```

```python
cap = stock.get_market_cap_by_ticker(date="20200929", market="ALL") 
cap = cap[['시가총액']]
cap
```

```python
all3 = all2.join(other=cap)
all3
```

```python
# 대형주 필터링
big = all3.sort_values(by='시가총액', ascending=False)[:200]
big
```

```python
big.sort_values(by='등락률', ascending=False)
```

```python
big_pct20 = big.sort_values(by='등락률', ascending=False)[:20]
big_pct20
```

```python
pct_df = stock.get_market_price_change_by_ticker(
    fromdate="20201006",  # 2020-10-05일 종가가 시가 컬럼임
    todate="20211008",
    market="ALL"
)
pct_df = pct_df[['시가', '종가', '등락률']]
pct_df
```

```python
big_mom_result = big_pct20.join(pct_df, lsuffix='_m')
big_mom_result
```

```python
big_mom_result['원금'] = 10000000
big_mom_result['매도금'] =  (big_mom_result['원금'] * (100 + big_mom_result['등락률'])) / 100
```

```python
최초자산 = big_mom_result['원금'].sum()
최종자산 = big_mom_result['매도금'].sum()
big_mom_cagr = (최종자산/최초자산) ** (1/1) - 1    # 12개월
big_mom_cagr * 100
```
