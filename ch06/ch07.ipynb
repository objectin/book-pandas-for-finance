## 7.1 Timestamp

```python
import pandas as pd

df = pd.read_excel("data/ss_ex_1.xlsx" , index_col=0)
df
```

```python
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df.head(3)
```

```python
print(df["거래량"].shift(1))
```

```python
df["전일거래량"] = df["거래량"].shift(1)
df[ ['거래량', '전일거래량'] ]
```

```python
df["전일거래량"] = df["거래량"].shift(1)
cond = df["거래량"] > df["전일거래량"]
df[cond]
```

```python
print("상승일:", len(df[cond]))
print("영업일:", len(df))
```

```python
df['D+1시가'] = df['시가'].shift(-1)
cond = df['종가'] < df['D+1시가']
print(len(df[cond]))
```

## 7.2 이동평균

```python
import pandas as pd

df = pd.read_excel("data/ss_ex_1.xlsx", index_col=0)
df.index = pd.to_datetime(df.index)
df = df.sort_index()[["종가"]]

df['종가D-1'] = df['종가'].shift(1)
df['종가D-2'] = df['종가'].shift(2)
df['ma3'] = (df['종가'] + df['종가D-1'] + df['종가D-2']) / 3
df
```

```python
df['rolling3'] = df['종가'].rolling(3).mean()
df.head()
```

```python
df = pd.read_excel("data/ss_ex_1.xlsx", index_col=0)
df.index = pd.to_datetime(df.index)

df['ma3'] = df['종가'].rolling(5).mean().shift(1)

cond = df['ma3'] < df['시가']
print("상승일:", len(df[cond]))
print("영업일:", len(df))
```

## 데이터 샘플링

```python
import pandas as pd

df = pd.read_excel("data/ss_ex_1.xlsx", index_col=0)
df.index = pd.to_datetime(df.index)
df = df.sort_index()[['시가', '저가', '고가', '종가', '거래량']]
df.head()
```

```python
df.resample('M').first()
```

```python
df.resample('MS').first()
```

```python
how = {
   "시가": "first",
   "종가": "last",
   "고가": max,
   "저가": min,
   "거래량": sum,
}

df.resample('MS').apply(how)
```

```python
df.resample('3D').apply(how).head()
```

```python
temp = df.resample('3D').apply(how)
temp.index = temp.index + pd.to_timedelta("2D")
temp.dropna().head()
```

```python
df.resample('3D', offset='1D').apply(how)
```

```python
df.resample('12H').apply(how)
```

```python
df.resample('12H').apply(how).ffill()
```

```python
df.resample('12H').apply(how).ffill()
```

```python
import numpy as np 

temp = df.resample('12H').apply(how)
cond = temp.isna().any(axis=1)
temp.loc[cond, '거래량'] = np.NaN
temp
```

```python
temp.ffill()
```
